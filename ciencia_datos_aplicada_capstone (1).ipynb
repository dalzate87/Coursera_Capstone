```python
# Usamos la autorización para hacer llamadas a la API de Foursquare en la version 3
# Generamos cada uno de los json a partir de las urls, los convertimos en diccionarios 
# y extraemos el nombre, latitud, longitud y categoria de lugares de interés a un radio 
# de 500 metros para cada vecindario

import json
headers = {
    "Accept": "application/json",
    "Authorization": "fsq3NMWGfjOVVhGqVteo4IZTw7TrzwJVJG8s0Fb0KFm3rlY="}

nombre = []
lat = []
lng = []
categoria = []


for url in urls: 
  response = requests.request("GET", url, headers=headers)
  dicc = json.loads(response.text)['results']
  for valor in range(0,len(dicc)):
    categoria.insert(valor, dicc[valor]['categories'][0]['name'])
    nombre.insert(valor, dicc[valor]['name'])
    lat.insert(valor, dicc[valor]['geocodes']['main']['latitude'])
    lng.insert(valor, dicc[valor]['geocodes']['main']['longitude'])
```

```python
datos_explorados = {'nombre':nombre, 'latitud':lat, 'longitud':lng, 'categoria':categoria}
```

```python
#Contamos los lugares explorados por categoría y seleccionamos las 10 categorias 
datos_explorados.groupby(['categoria']).count()['nombre'].sort_values(ascending = False).iloc[:10,]
```

```python
# Codificación de las categorias
dummy_categories = pd.get_dummies(datos_explorados[['categoria']], prefix = "", prefix_sep="")
dummy_categories.head()

```

```python
# Unimos los datos explorados junto con la codificación para la variable categoria del lugar
datos_explorados = datos_explorados.join(dummy_categories)
datos_explorados.head(5)
```

```python
# Número de agrupaciones
k_clusters = 5
datos_kmeans = datos_explorados.drop(['nombre','categoria'], axis = 1)
```

```python
datos_kmeans.head(5)
```

```python
# Ejecutar k means
kmeans = KMeans(n_clusters=k_clusters,random_state = 0).fit(datos_kmeans)

# Revisar las etiquetas
kmeans.labels_[0:11]
grupos = pd.DataFrame({'label':kmeans.labels_})


# Unir los grupos a los datos
datos_kmeans = datos_kmeans.join(grupos)
datos_kmeans = datos_kmeans.join(datos_explorados[['nombre','categoria']])
datos_kmeans.head()
```

```python
# Crear mapa de North York - Canadá
mapa_clusters = folium.Map(location = [43.7668100,-79.4163000], zoom_start=11)

# Construir esquema de color
x = np.arange(k_clusters)
ys = [i + x + (i*x)**2 for i in range(k_clusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# Añadir marcadores al mapa
markers_colors = []
for lat, lng, poi, cluster in zip(datos_kmeans['latitud'], datos_kmeans['longitud'], datos_kmeans['nombre'], datos_kmeans['label']):
  label = folium.Popup(str(poi) + 'Cluster' + str(cluster), parse_html = True)
  folium.CircleMarker(
      [lat,lng],
      radius = 5,
      popup = label,
      color = rainbow[cluster-1],
      fill = True,
      fill_color = rainbow[cluster-1],
      fill_opacity = 0.7).add_to(mapa_clusters)
mapa_clusters
  
```
